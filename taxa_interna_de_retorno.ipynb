```python
# Vamos inicializar as variáveis
projeto_A = np.array([-10000, 5000, 4000, 3000])
projeto_B = np.array([-30000, 9000, 12000, 15000])
projeto_C = np.array([-50000, 30000, 10000, 20000])
```

```python
irr_projeto_a = npf.irr(projeto_A)
irr_projeto_b = npf.irr(projeto_B)
irr_projeto_c = npf.irr(projeto_C)
```

```python
print(f'IRR do projeto A: {round(irr_projeto_a, 4)}')
print(f'IRR do projeto B: {round(irr_projeto_b, 4)}')
print(f'IRR do projeto C: {round(irr_projeto_c, 4)}')
```
